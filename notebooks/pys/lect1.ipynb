***Lists:***


* Lists store multiple values in order.

* nums[0] and nums[-1] work like string indexing.

* append() adds to the end.

* for x in nums: iterates through each item.

* In the REPL, for x in nums: print(x) is valid, though the multi-line indented form is the more common style.
Then
* Then try `len()` `insert()` `pop()` `remove()` and `range(len(...))`


```python
nums = [10, 20, 30, 40]
print(nums)
print(nums[0])
print(nums[-1])
print(nums[1:3])

nums.append(50)
nums.append(60)
print(nums)

print("----------------")
for x in nums:
    print(x)
```

```python
fruits = ["mango", "apple", "banana", "grape"]

print(len(fruits))

fruits.insert(1, "orange")
print(fruits)

fruits.pop()
print(fruits)

fruits.remove("apple")
print(fruits)

for i in range(len(fruits)):
    print(i, fruits[i])
```

***Dictionaries*** : Dictionaries store data as key-value pairs, and you access values by key instead of numeric position.

* Dictionaries let you add, update, and safely read values with methods like `get()`, `update()`, `pop()`, and `items()`.

* `get()` returns a value if the key exists, and a safe fallback behavior when it does not.

* `pop()` removes a key and returns its value.

* `items()` is the clean way to loop over both keys and values together.



```python
student = {
    "name": "Aadi",
    "age": 23,
    "city": "Ahwa"
}

print(student)
print(student["name"])
print(student.get("age"))
print(student.keys())
print(student.values())
print(student.items())

print("---------------------")
student["age"] = 26
student["country"] = "India"
print(student)

print("---------------------")

student["age"] = 26
student["country"] = "India"

print(student)
print(student.get("name"))
print(student.get("grade"))
print(student.pop("city"))
print(student)

print("---------------------")
for k, v in student.items():
    print(k, "->", v)
```

***Sets:***  A Python set is an unordered collection of unique items, so duplicates vanish automatically, and you use membership tests and set operations instead of indexing

* Sets do not allow duplicate values, add() inserts one element, and remove() deletes an element. Membership testing with in is one of the strongest reasons sets are useful.

* There are union, intersection, difference, and symmetric difference, respectively.


```python
s = {1, 2, 3, 3, 4}
print(s)

s.add(5)
print(s)

s.remove(2)
print(s)

print(3 in s)
print(9 in s)

print("-----------------")
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a | b)
print(a & b)
print(a - b)
print(a ^ b)
```

***Functions:*** Functions bundle reusable logic, can take parameters, and can return values; that’s the point where your code stops being a pile of commands and starts becoming organized.

*

```python
def greet(name):
    print("Hello,", name)

greet("Aadi")
greet("Python")

print("-----------------")

def add(a, b):
    return a + b

result = add(7, 5)
print(result)

print("-----------------")

def power(base, exp=2):
    return base ** exp

print(power(4))
print(power(4, 3))

print("-----------------")

# Functions with default arguments
def greet(name, msg="Good morning"):
    return msg + ", " + name + "!"

print(greet("Aadi"))
print(greet("Aadi", "Good afternoon"))

print("-----------------")

# Function returning multiple values
def min_max(nums):
    return min(nums), max(nums)

result = min_max([3, 7, 1, 9, 4])
print(result)
```
