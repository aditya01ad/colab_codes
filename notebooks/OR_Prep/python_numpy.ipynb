## Array Creation

```python
## np.array, np.zeros, np.ones, np.arange, np.linspace

# array
a = np.array([1, 2, 3, 4, 5])
print(a)
print(a * 2)       # vectorized — no loop needed
print(a ** 2)
print(np.sum(a))
print(np.mean(a))

print("------------------------------")
# 2D array
M = np.array([[1, 2, 3],
              [4, 5, 6]])

print(M.shape)     # (2, 3)
print(M[0])        # first row
print(M[:, 1])     # second column
print(M.T)         # transpose

print("------------------------------")
# Some matrises
print(np.zeros((3, 3)))
print((np.ones((2, 4))))
print(np.eye(3))
print(np.arange(0, 10, 2))
print(np.linspace(0, 1, 5))

print("------------------------------")
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print(A + B)
print(A * B)
print(A @ B)
print(np.linalg.det(A))
print(np.linalg.eig(A))
```

## Indexing & Reshaping

```python
# Slicing, fancy indexing, reshape
a = np.array([1, 2, 3])
print(a + 10)

A = np.array([[1, 2, 3],
              [4, 5, 6]])
b = np.array([10, 20, 30])
print(A + b)

x = np.array([[1],
              [2],
              [3]])
y = np.array([10, 20, 30])
print(x + y)


a = np.array([1, 2, 3])
print(a.shape)   # (3,)

M = np.array([[1, 2, 3],
              [4, 5, 6]])
print(M.shape)   # (2, 3)


x = np.array([1, 2, 3])
y = np.array([[1], [2], [3]])
print(x.shape)
print(y.shape)
print(x + y)
```

Shape intuition
> `a + 10` works because scalar behaves like shape ().

> `A + b` works because b is treated like [10, 20, 30] across each row.

> `x + y` gives a full 3 x 3 result because shapes `(3, 1)` and `(3,)` broadcast to `(3, 3)`.

> A 1D array like `[1, 2, 3]` has shape `(3,)`, meaning 3 elements in one line.

> A 2D array like `[[1, 2, 3], [4, 5, 6]]` has shape `(2, 3)`, meaning 2 rows and 3 columns.

> A 3D array would have three numbers in its shape tuple, one for each axis.



```python
a = np.array([1, 2, 3])        # shape (3,)
b = np.array([[1, 2, 3]])      # shape (1, 3)
c = np.array([[1], [2], [3]])   # shape (3, 1)

print("a.T:", a.T)
print("b.T:", b.T)
print("c.T:", c.T)
```

> In NumPy, a vector is usually treated as a 1D array, while a matrix is explicitly 2D, so a, b, and c are intentionally different shapes.

> [1, 2, 3] in NumPy is a vector.

> [[1], [2], [3]] is a 3×1 column matrix.

> [[1, 2, 3]] is a 3×1 row matrix.

> NumPy treats a as a 1D vector, while b and c are 2D matrices with orientation, so transpose changes the 2D ones but leaves the 1D one unchanged.

> a.T stays [1 2 3].

> b.T becomes a column shape [[1], [2], [3]].

> c.T becomes a row shape [[1 2 3]]
