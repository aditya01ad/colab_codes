## Array Creation

```python
## np.array, np.zeros, np.ones, np.arange, np.linspace

# array
a = np.array([1, 2, 3, 4, 5])
print(a)
print(a * 2)       # vectorized — no loop needed
print(a ** 2)
print(np.sum(a))
print(np.mean(a))

print("------------------------------")
# 2D array
M = np.array([[1, 2, 3],
              [4, 5, 6]])

print(M.shape)     # (2, 3)
print(M[0])        # first row
print(M[:, 1])     # second column
print(M.T)         # transpose

print("------------------------------")
# Some matrises
print(np.zeros((3, 3)))
print((np.ones((2, 4))))
print(np.eye(3))
print(np.arange(0, 10, 2))
print(np.linspace(0, 1, 5))

print("------------------------------")
```
