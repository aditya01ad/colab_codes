## Variables & Operators

```python
## Variables and basic operators
# Who are you?
name = "Aditya"
age = 22
is_mathematician = True

print(f"Name: {name}, Age: {age}, Mathematician: {is_mathematician}")
print(type(name), type(age), type(is_mathematician))

print(isinstance(True, int))   # True or False?
print(isinstance(True, bool))  # True or False?
print(True + True + True)      # what does this give?
print(type(True + 0))          # still bool?

print("-------------------------------")

x = 10
y = 3
print("x + y = ", x + y)
print("x - y = ", x - y)
print("x * y = ", x * y)
print("x / y = ", x / y)
print("x // y = ", x // y)
print("x ** y = ", x ** y)
print("x % y = ", x % y)

print("-------------------------------")


```

```python
class MyInt(int):
    pass

x = MyInt(1)
print(isinstance(x, int))
print(isinstance(x, MyInt))
print(x + 2)
print(type(x + 2))
print(type(x))

class MyInt(int):
    def __add__(self, other):
        return MyInt(super().__add__(other))  # explicitly wrap

x = MyInt(1)
print(isinstance(x, int))
print(isinstance(x, MyInt))
print(x + 2)
print(type(x + 2))
print(type(x))
```

## Strings & Slicing

```python
## String operations

# - - - - - Strings - - - - -
s = "Optimization"

print(len(s))          # 12 — number of characters
print(s[0])            # 'O' — first character
print(s[-1])           # 'n' — last character
print(type(s))         # <classStrings

print("-----------------------------------")

# - - - - - Slicing - - - - -
print(s[0:4])
print(s[4:])
print(s[:4])
print(s[::2])
print(s[::-1])
print(s[2:8:2])

print("-----------------------------------")

# - - - - Common String Methods - - - -
s = "  hello world  "

print(s.upper())
print(s.lower())
print(s.replace("world", "Aditya"))
print(s.split())

print("-".join(['a', 'b', 'c']))
print("world" in s)
print(s.strip().startswith("hello"))
print(s.count("l"))


print("-----------------------------------")

# - - - f-strings - - - (we'll use this everywhere)
city = "Bhubaneswar"
temp = 38.5

print(f"It is {temp:.1f}°C in {city}")
print(f"2 + 2 = {2 + 2}")
print(f"{'hello':>10}")

print("-----------------------------------")

s = "hello"
#s[0] = "H" -> TypeError: 'str' object does not support item assignment
s = "H" + s[1:]
print(s)

print("-----------------------------------")

line = "Delhi,Mumbai,Chennai,Kolkata"
cities = line.split(",")
print(cities)
print(len(cities))
print("->".join(cities))
```

## Lists, Dicts, Sets, Tuples

## The key difference is:

* List: ordered, mutable.

* Tuple: ordered, immutable.

* Set: unordered, unique elements only.

* Dictionary: key-value mapping, mutable.

```python
nums = [10, 20, 30, 40]
print(nums)
print(type(nums))
print(nums[0])
print(nums[-1])
print(nums[1:3])

nums.append(50)
print(nums)

nums.insert(1, 15)
print(nums)

nums.pop()
print(nums)

nums.remove(20)
print(nums)
```

```python
point = (3, 5)
print(point)
print(type(point))
print(point[0])
print(point[1])

x, y = point
print(x)
print(y)
```

```python
s = {1, 2, 3, 3, 4}
print(s)
print(3 in s)
print(9 in s)
s.add(5)
print(s)
```

```python
student = {"name": "Aadi", "age": 25}
print(student)
print(student["name"])
print(student.get("age"))
student["city"] = "Ahwa"
print(student)
```
