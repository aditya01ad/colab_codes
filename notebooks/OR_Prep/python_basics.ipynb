## Variables & Operators

```python
## Variables and basic operators
# Who are you?
name = "Aditya"
age = 22
is_mathematician = True

print(f"Name: {name}, Age: {age}, Mathematician: {is_mathematician}")
print(type(name), type(age), type(is_mathematician))

print(isinstance(True, int))   # True or False?
print(isinstance(True, bool))  # True or False?
print(True + True + True)      # what does this give?
print(type(True + 0))          # still bool?

print("-------------------------------")

x = 10
y = 3
print("x + y = ", x + y)
print("x - y = ", x - y)
print("x * y = ", x * y)
print("x / y = ", x / y)
print("x // y = ", x // y)
print("x ** y = ", x ** y)
print("x % y = ", x % y)

print("-------------------------------")


```

```python
class MyInt(int):
    pass

x = MyInt(1)
print(isinstance(x, int))
print(isinstance(x, MyInt))
print(x + 2)
print(type(x + 2))
print(type(x))

class MyInt(int):
    def __add__(self, other):
        return MyInt(super().__add__(other))  # explicitly wrap

x = MyInt(1)
print(isinstance(x, int))
print(isinstance(x, MyInt))
print(x + 2)
print(type(x + 2))
print(type(x))
```
