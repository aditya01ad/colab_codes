## Lists, Dicts, Sets, Tuples

* List: ordered, mutable.

* Tuple: ordered, immutable.

* Set: unordered, unique elements only.

## Conditionals

````python
## if / elif / else\

# Basic if/elif/else
x = 42

if x > 100:
  print("large")
elif x > 10:
  print("medium")
else:
  print("small")

# logical operators
age = 20
has_id = True

if age >= 18 and has_id:
  print("Entry allowed")
else:
  Print("Entry denied")

score = 45
if score < 0 or score > 100:
  Print("Invalid score")
else:
  print(f"Score : {score}")

is_closed = False
if not is_closed:
  print("Shop is open")

# Nested conditions
marks = 78

if marks >= 90:
  print("Grade: A")
elif marks >= 75:
  if marks >=80:
    print("Grade: B+")
  else:
    print("Grade: B")
elif marks >=60:
  print("Grade: C")
else:
  print("Grade: F")

# One-liner(ternary expression)
x = 15
label = "odd" if x%2!=0 else "even"
print(label)

# OR-relevant
demand = 120
capacity = 100

if demand > capacity:
  shortage = demand - capacity
  print(f"Shortage: {shortage} units")
elif demmand == capacity:
  print("perfactly balanced")
else:
  suplus = capacity - demmand
  print(f"Surplus: {surplus} units")
````

## Loops

````python
## for and while loops

# For
for i in range(5):
  print(i)
print("------------------")
for i in range(2,10):
  print(i)
print("------------------")
for i in range(1,10,2):
  print(i)
print("------------------")
````

> `range()` can take 1, 2, or 3 arguments, and it stops before the end value.

> `break` exits the loop completely, while `continue` skips only the current iteration.



````python
# while
count = 1
while count <= 5:
  print(count)
  count += 1

print("------------------")

# break
for i in range(10):
  if i == 5:
    break
  print(i)

print("------------------")

# continue
for i in range(10):
  if i == 5:
    continue
  print(i)

````

> In OR-style scripts: counting, accumulating, iterating over rows, and building totals.



````python
# OR-block
times = 0
total = 0

while times < 5:
  total += times
  times += 1

print(total)
````

````python
for i in range(1, 21):
  if i%3 == 0:
    continue
  print(i)
````

## Functions

### syntax:


```
def function_name(parameters):
    # body
    return value
```



````python
## Functions, default args, multiple return values

# greet
def greet(name):
  return "Hello, " + name

print(greet("Aadi"))

# square
def square(n):
  return n**2
x = 7
print(f"{x} square is {square(x)}")

# even ?
def is_even(n):
  return n % 2 == 0
x = 7
print(f"{x} is even? {is_even(x)}")

# factorial
def factorial(n):
  fact = 1
  for k in range(1, n+1):
    fact = fact * k
    print(k)
  return fact
print("factorial:", factorial(0))
````

````python
def stats(numbers, show_range=False):
    if not numbers:
        raise ValueError("numbers must not be empty")
    n = len(numbers)
    mean = sum(numbers) / n
    variance = sum((x - mean) ** 2 for x in numbers) / n
    if show_range:
        range = max(numbers) - min(numbers)
        return mean, variance, range
    return mean, variance
print(stats([1, 2, 3]))
print(stats([1, 2, 3], show_range=True))
````

## List Comprehensions

**Syn:**



```
[expression for item in iterable if condition]
```



````python
## List comprehensions

# Rewrite the this loop using a list compresension

# loop
squares = []
for x in range(1, 21):
  squares.append(x*x)
print(squares)
# comp
squares = [ x*x for x in range(1,11)]
print(squares)

# comp
evens = [x for x in range(1,30) if x%2==0]
print(evens)
words = ["cat", "dog", "elephant", "ant"]
lengths = [len(w) for w in words]
print(lengths)
````

````python
# Dictionary examples
square_map = {x: x*x for x in range(1,9)}
print(square_map)

nums= [1,2,3,4,5]
even_squares = {x : x*x for x in nums if x%2 == 0}
print(even_squares)
````

````python
# Lambda
add_10 = lambda x: x + 10
print(add_10(27))

multiply = lambda a,b: a*b
print(multiply(8,47))

nums = [1, 2, 3, 4]
doubled = list(map(lambda x: x*2, nums))
print(doubled)
````

````python
# enumerate()
names = ["Aadi", "Aria", "Lyra"]
for i, name in enumerate(names):
    print(i, name)

for i, name in enumerate(names, start=1):
    print(i, name)

# zip()
names = ["aadi", "aria", "lyra"]
scores = [90, 95, 88]

for name, score in zip(names, scores):
  print(name, score)

pairs = [(name, score) for name, score in zip(names, scores)]
print(pairs)
````

````python
# Nested comprehension
matrix = [
    [1, 2], [3, 4], [5,6]
]
flat = [x for row in matrix for x in row]
print(flat)

grid = [[i * j for j in range(1, 4)] for i in range(1, 4)]
print(grid)
````

## File I/O

> Use open() with a mode like "r", "w", or "a", and the safest pattern is with open(...) as f: so the file closes automatically.

````python
# Reading and writing files

# read
with open("sample.txt", "r") as f:
    text = f.read()
print(text)
print("----------------------------")
# write
with open("sample.txt", "w") as f:
    f.write("Hello, Aadi\n")
    f.write("Python file I/O")
# read
with open("sample.txt", "r") as f:
    text = f.read()
print(text)
print("----------------------------")

# append
with open("sample.txt", "a") as f:
    f.write("\nAdded later")
# read
with open("sample.txt", "r") as f:
    text = f.read()
print(text)
print("----------------------------")

# read line by line
with open("sample.txt", "r") as f:
    for line in f:
        print(line.strip())
print("----------------------------")
````
